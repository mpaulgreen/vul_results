```python
min_value = 0
max_value = min_value + 35
runs = 6
```

```python
# Path to the local Excel file
file_path = "experiment_data_fixed.xlsx" 
# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file_path)
```

#### Send request to LLM

```python
import time
# send request to ai agent
for i in range (min_value, max_value):
    row = df.iloc[i]
    print(row["request_id"])
    request_cve.process_scan(
        row["request_id"] , 
        runs,
        "registry.redhat.io/"+row["package"],
        row["CVE"]
    )
    # Enable this to clean up the data as agent morpheus crashes for some data and 
    # when the embedding takes too long, it loses following source messages. Need to investigate it later.
    time.sleep(60) # Typically on first run
    
```

```python
runs_file = "experiment_data_fixed_runs.xlsx"
if os.path.exists(runs_file):
    output_df = pd.read_excel(runs_file)
else:
    output_df = df
```

```python
for i in range (min_value, max_value):
    row = output_df.iloc[i]
    _id = row["request_id"] + str(runs)
    status, label = response_cve.extract_justification(response_cve.find_document_by_scan_id(response_cve.collection, _id))
    output_df.at[i, ("status" + str(runs))] = status
    output_df.at[i, ("label" + str(runs))] = label
    output_df.at[i, ("cve_request_id" + str(runs) )] = _id
```
